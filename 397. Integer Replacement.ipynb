```python
class Solution(object):
    def integerReplacement(self, n):
        dp = [n+1]*(n+1)
        
        dp[0]=0
        dp[1]=0
        
        for i in range(2,n+1):
            if i%2 == 0:
                dp[i] = min(dp[i], 1+dp[int(i/2)])
            else:
                dp[i] = min(dp[i], 2+dp[int((i-1)/2)], 2+dp[int((i+1)/2)])
            print(f"{i}: {dp[i]}")
        return dp[n]
    
s = Solution()
s.integerReplacement(100)
```
