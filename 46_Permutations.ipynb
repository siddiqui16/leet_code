```python
class Solution(object):
    def permute(self, nums, d = 0):

      if len(nums) < 2:
        return [nums]
      else:
        p = []
        for i,n in enumerate(nums):
          res = self.permute(nums[:i] + nums[i+1:], d+1)
          p.extend([[n]+j for j in res])
        if d == 0:
          print(len(nums), len(p))
        return p


s = Solution()
s.permute([1,2,3,4,5,6])
```
