```python
class Solution:
    def fullJustify(self, words, maxWidth):

        def get_lengths(words):
            len_ = []
            for w in words:
                len_.append(len(w))
            return len_

        lengths = get_lengths(words)

        r = 0
        line = list()
        seeds = list()
        for w, l in zip(words, lengths):
            if r + l > maxWidth:
                seeds.append(line)
                line = list()
                r = 0
            line += [w]
            r += l + 1
        seeds.append(line)

        justified = []
        for i, line_s in enumerate(seeds):
            g = len(line_s) - 1
            s = maxWidth - (sum(get_lengths(line_s)) + g)

            if i+1 == len(seeds):
                spaces = [0] * (g+1)
                spaces[-1] = s
            else:
                if g > 0:
                    dist = int(s/g)
                    spaces = [dist] * g
                    for i in range(s%g):
                        spaces[i] += 1
                    spaces = spaces + [0]
                else:
                    spaces = [s]

            j = []
            for w_, p in zip(line_s, spaces):
                j+= [f"{w_}{' ' * p}"]
            j_str = " ".join(j)
            justified.append(j_str)
            print(f"{j_str} : {len(j_str)} > {sum(spaces)} {round(sum(spaces)/len(j_str)*100 + 0.01, 2)}")
        print("-"*80)

        return justified




s = Solution()

print(s.fullJustify(words = ["This", "is", "an", "example", "of", "text", "justification."], maxWidth = 16))

print(s.fullJustify(words = ["What","must","be","acknowledgment","shall","be"], maxWidth = 16))

print(s.fullJustify(words = ["Science","is","what","we","understand","well","enough","to","explain","to","a","computer.","Art","is","everything","else","we","do"], maxWidth = 20))

print(s.fullJustify(words = ["Lorem","ipsum","dolor","sit","amet","consectetur","adipiscing","elit","Nulla","hendrerit","luctus","elit","ac","lacinia","Praesent","risus","lorem","faucibus","sed","dolor","id","dignissim","dictum","velit","Sed","ut","viverra","magna","Ut","cursus","nulla","quam","et","rhoncus","justo","viverra","tempus","Pellentesque","porttitor","vel","dolor","et","facilisis","Cras","porta","est","vitae","posuere","blandit","magna","ligula","vehicula","augue","et","tincidunt","elit","risus","mollis","purus","Nunc","eu","enim","sed","neque","vestibulum","venenatis","eget","sed","enim","Donec","ultrices","lectus","non","gravida","ultrices","tellus","arcu","hendrerit","erat","vel","placerat","orci","odio","eget","est","Mauris","in","nunc","vitae","massa","fermentum","mollis","Sed","lobortis","nibh","eget","condimentum","porttitor","Ut","venenatis","metus","vel","ex","condimentum","sit","amet","faucibus","eros","fermentum","Sed","sit","amet","feugiat","mauris","Sed","dictum","turpis","in","eros","rutrum","tincidunt","Quisque","ac","lobortis","augue","Aliquam","et","sollicitudin","sapien","Maecenas","nulla","elit","bibendum","ut","urna","id","volutpat","imperdiet","sem","Vestibulum","condimentum","iaculis","turpis","vel","sagittis","Etiam","vitae","elit","lacinia","justo","tristique","elementum","non","eget","nisi","Vestibulum","scelerisque","augue","eu","maximus","fermentum","Aliquam","erat","volutpat","In","elementum","diam","vitae","tortor","condimentum","in","auctor","dolor","venenatis","Sed","justo","nibh","imperdiet","quis","tincidunt","fringilla","interdum","a","lorem","Nam","massa","leo","ornare","a","quam","quis","posuere","accumsan","lorem","Curabitur","ut","risus","iaculis","commodo","tellus","ut","lacinia","mi","In","egestas","et","sem","nec","gravida","Aliquam","erat","lacus","convallis","at","commodo","ac","maximus","et","est","Interdum","et","malesuada","fames","ac","ante","ipsum","primis","in","faucibus","Praesent","egestas","efficitur","semper","Proin","sodales","ligula","vel","augue","lacinia","vitae","semper","diam","lobortis","Fusce","enim","urna","vestibulum","eget","ligula","id","tincidunt","venenatis","eros","Morbi","in","turpis","sit","amet","nisi","porta","egestas","Nunc","eget","urna","sem","Quisque","ullamcorper","tortor","sed","consequat","elementum","Vivamus","scelerisque","porttitor","vestibulum","Morbi","libero","justo","finibus","nec","eros","eget","placerat","luctus","diam","Cras","commodo","laoreet","diam","et","imperdiet","Phasellus","ullamcorper","vestibulum","orci","sit","amet","varius","nisi","consequat","et","Nam","sem","dui","vulputate","quis","odio","at","sagittis","placerat","eros","Nunc","mi","nulla","aliquet","eu","tempor","et","volutpat","blandit","diam","Praesent","pharetra","hendrerit","nibh","ultrices","semper","est","dictum","malesuada","Mauris","non","ligula","eget","sem","efficitur","dignissim","id","vel","neque","Quisque","dolor","mauris","mollis","vitae","dolor","quis","laoreet","consequat","tortor","Praesent","sagittis","vestibulum","ante","a","semper","Ut","blandit","turpis","ac","nibh","sodales","non","porttitor","turpis","suscipit","Vivamus","hendrerit","at","ligula","id","mattis","Proin","diam","nisi","scelerisque","consectetur","egestas","et","semper","et","ligula","Fusce","eu","massa","a","lectus","fermentum","blandit","non","et","diam","Curabitur","faucibus","ornare","nisl","a","euismod","Fusce","pretium","nisi","vel","arcu","euismod","quis","ultrices","enim","viverra","Pellentesque","habitant","morbi","tristique","senectus","et","netus","et","malesuada","fames","ac","turpis","egestas","Donec","facilisis","urna","arcu","vel","fringilla","nisl","tincidunt","a","Pellentesque","finibus","tincidunt","orci","in","euismod","neque","viverra","non","Integer","volutpat","semper","neque","vel","maximus","leo","mattis","ut","Sed","at","mollis","elit","Nulla","commodo","mattis","odio","nec","efficitur","nunc","accumsan","quis","Vestibulum","et","enim","eu","dui","porta","suscipit","Phasellus","vitae","nibh","faucibus","venenatis","dolor","a","efficitur","quam","Aliquam","quis","velit","dolor","Quisque","vitae","dignissim","ex","Pellentesque","vulputate","rhoncus","sapien","sit","amet","aliquet","nisi","viverra","sed","Sed","sapien","ipsum","fringilla","in","arcu","vel","rhoncus","accumsan","lorem","Vestibulum","porta","ante","in","gravida","mollis","quam","ex","dapibus","nibh","ultricies","condimentum","velit","urna","quis","libero","Nam","blandit","vel","risus","eget","scelerisque","Nullam","id","sollicitudin","mauris","eu","vehicula","est"],
                    maxWidth = 80))
```
