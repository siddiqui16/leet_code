```python
class Solution:
    def maxPoints(self, points) -> int:
        unique_pts = set()
        for p in points:
            unique_pts.add("%s" % p)

        if len(unique_pts) <= 2:
            return len(points)

        def add_points(p):
            return p[0]+p[1]

        def sub_points(p):
            return p[0]-p[1]

        def horizontal(p):
            return p[1]

        def vertical(p):
            return p[0]


        l_add_points = [r for r in map(add_points, points)]
        l_sub_points = [r for r in map(sub_points, points)]
        l_horizontal = [r for r in map(horizontal, points)]
        l_vertical = [r for r in map(vertical, points)]

        from collections import Counter
        m_add = Counter(l_add_points)
        m_sub = Counter(l_sub_points)
        m_hor = Counter(l_horizontal)
        m_ver = Counter(l_vertical)

        print(m_add)
        print(m_sub)
        print(m_hor)
        print(m_ver)

        max_points = 0
        for v in m_add.values():
            if v > max_points:
                max_points = v

        for v in m_sub.values():
            if v > max_points:
                max_points = v

        for v in m_hor.values():
            if v > max_points:
                max_points = v

        for v in m_ver.values():
            if v > max_points:
                max_points = v

        return max_points



s =Solution()
print(s.maxPoints([[1,1],[6,2]]))
print(s.maxPoints([[1,1],[2,2],[3,3]]))
print(s.maxPoints([[1,1],[3,2],[5,3],[4,1],[2,3],[1,4]]))


print(s.maxPoints([[2,3],[3,3],[-5,3]]))
print(s.maxPoints([[4,0],[4,-1],[4,5]]))
print(s.maxPoints([[1,1],[1,1],[2,3]]))

```
