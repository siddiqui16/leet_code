```python
class Solution(object):
    def generate(self, numRows):
        def rsum(x):
            q = []
            for j in range(1,len(x)):
                q.append(x[j-1]+x[j])
            return q
        
        r = []
        for i in range(numRows):
            if i<=1:
                l = [1] * (i+1)
            else:
                l = [1] + rsum(r[-1]) + [1]
            r.append(l)
        return r
        
        
s = Solution()
s.generate(20)
```
