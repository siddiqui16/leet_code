# Solution 1
`complexity = O(n^3)`

```python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        l = set()
        n = len(nums)
        
        for i in range(n-2):
            for j in range(i+1, n-1):
                for k in range(j+1, n):
                    if nums[i] + nums[j] + nums[k] == 0:
                        print(i,j,k)
                        x = sorted([nums[i], nums[j], nums[k]])
                        l.add(",".join([f"{x_}" for x_ in x]))
        ans = []
        for t in l:
          ans.append([int(k) for k in t.split(",")])
        return ans

s = Solution()
s.threeSum(nums = [-1,0,1,2,-1,-4])
```

# Solution 2
`complexity = O(n^2 + nm)`

```python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """
        sum2 = dict()
        n = len(nums)
        l = set()
        
        for i in range(n-1):
            for j in range(i+1, n):
              s2 = nums[i] + nums[j]
              if s2 not in sum2:
                sum2[s2] = []
              sum2[s2].append([i,j])

        for k in range(n):
          s2 = 0 - nums[k]
          if s2 in sum2:
            for i,j in sum2[s2]:
              if i!=k and j!=k:
                l.add(tuple(sorted([nums[i],nums[j],nums[k]])))

        print(sum2)
        ans = []
        for t in l:
          ans.append(list(t))
        return ans

s = Solution()
#s.threeSum(nums = [-1,0,1,2,-1,-4])
s.threeSum(nums = [-1,0,1,2,-1,-4])
```
