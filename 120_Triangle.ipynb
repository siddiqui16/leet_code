# Bottom Up

```python
class Solution:
  p = {}
  def print(self, t):
    for l in t:
      print(' '.join([f"{v:2}" for v in l]))
    print("-"*40)

  def minimumTotal(self, triangle, r=0, c=0):
    self.print(triangle)
    n = len(triangle)
    for r in range(n-2,-1,-1):
      for c in range(r+1):
        triangle[r][c] += min(triangle[r+1][c], triangle[r+1][c+1])
    self.print(triangle)

    return triangle[0][0]
          
    


s = Solution()

inputs = [
  #([[2],[3,4],[6,5,7],[4,1,8,3]], 11),
  #([[-10]], -10),
  ([[7],[-5,9],[6,5,2],[-8,-2,-7,3],[-2,6,-6,-1,4]],-6)
]

for i,o in inputs:
  print(o, s.minimumTotal(i))

```

# Top Down

```python
class Solution:
  p = {}
  def print(self, t):
    print("-"*40)
    for l in t:
      print(' '.join([f"{v:2}" for v in l]))

  def minimumTotal(self, triangle, r=0, c=0):
    self.print(triangle)
    n = len(triangle)
    for r in range(n):
      for c in range(r+1):
        #print((r,c), triangle[r][c], '-'*20)
        s = None
        for i,j in [(-1,0),(-1,-1)]:
          r_,c_ = r+i,c+j
          if -1<r_ and -1<c_<r_+1:
            if not s:
              s = triangle[r_][c_]
            else:
              s = min(s, triangle[r_][c_])
            #print("\t>", (r_,c_), s)
        if s:
          #print("\t+", (r,c), s, "+", triangle[r][c])
          triangle[r][c]+=s
    self.print(triangle)
    return min(triangle[-1])
          
    


s = Solution()

inputs = [
  #([[2],[3,4],[6,5,7],[4,1,8,3]], 11),
  #([[-10]], -10),
  ([[7],[-5,9],[6,5,2],[-8,-2,-7,3],[-2,6,-6,-1,4]],-6)
]

for i,o in inputs:
  print(o, s.minimumTotal(i))

```
