```python
import collections
class Solution:
    def deleteAndEarn(self, nums, d=0):
        if d==0:
            self.p = {}
        if len(nums) == 0:
            return 0
        if len(nums) == 1:
            return nums[0]
        
        s = []
        count_i = collections.Counter(nums)
        for i in sorted(count_i):
            jums = [nums[j] for j in range(len(nums)) if abs(i-nums[j]) > 1]
            count_j = collections.Counter(jums)
            j_str = '.'.join([f"{x}" for x in sorted(count_j)])
            if d<1:
                print(f"{d}:{i:4}  ({len(self.p):4}) -{'-'*(d*4+1)} {i}: {j_str}")
            if j_str not in self.p:
                self.p[j_str] = self.deleteAndEarn(jums, d+1)
            s.append(i*count_i[i] + self.p[j_str])
        return max(s)
    
s = Solution()
#s.deleteAndEarn([3,4,3])
#s.deleteAndEarn([2,2,3,3,3,4])
s.deleteAndEarn([8,3,4,7,6,6,9,2,5,8,2,4,9,5,9,1,5,7,1,4])
#s.deleteAndEarn([1,8,5,9,6,9,4,1,7,3,3,6,3,3,8,2,6,3,2,2,1,2,9,8,7,1,1,10,6,7,3,9,6,10,5,4,10,1,6,7,4,7,4,1,9,5,1,5,7,5])
#s.deleteAndEarn([12,32,93,17,100,72,40,71,37,92,58,34,29,78,11,84,77,90,92,35,12,5,27,92,91,23,65,91,85,14,42,28,80,85,38,71,62,82,66,3,33,33,55,60,48,78,63,11,20,51,78,42,37,21,100,13,60,57,91,53,49,15,45,19,51,2,96,22,32,2,46,62,58,11,29,6,74,38,70,97,4,22,76,19,1,90,63,55,64,44,90,51,36,16,65,95,64,59,53,93])
```
